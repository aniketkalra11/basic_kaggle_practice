**Generating synthetic dataset**

```python
def number_to_alphabet(number:int):
    alphabets = 'abcdefghijklmnopqrstuvwxyz'
    assert len(alphabets) == 26
    indx = number % 26
    assert indx < 26
    return alphabets[indx]
```

```python
number_rows = 1000000
row1 = [np.random.randint(1, 10000) for _ in range(number_rows)]
# print(row1)
row2 = [np.random.randint(1, number_rows) for _ in range(number_rows)]
# print(row2)
row3 = [number_to_alphabet(np.random.randint(1, number_rows)) for _ in range(number_rows)]
# print(row3)
dict_rows = {
    'row1': row1,
    'row2': row2,
    'row3': row3
}
df = pd.DataFrame(dict_rows)
```
