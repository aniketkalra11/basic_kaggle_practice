```python
from plotnine import *
from plotnine.data import mtcars
from ggplot import *
import  matplotlib.pyplot  as plt
```

```python
mtcars.head()
```

```python
plt.scatter(x=mtcars['wt'], y= mtcars['mpg'])
```

2D visalization using ggplot

```python
(ggplot(mtcars, aes('wt', 'mpg')) + geom_point() + theme_bw())
       
```

3D points

```python
ggplot(mtcars, aes('wt', 'mpg', color='factor(gear)')) + geom_point() + theme_bw()
```

4D points


```python
(ggplot(mtcars, 
        aes('wt', 'mpg', 
            color='factor(gear)', size='cyl')) + geom_point() + theme_bw())
```
