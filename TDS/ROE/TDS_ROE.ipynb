```python
from bs4 import BeautifulSoup
import requests
```

```python
response = requests.get("https://www.imdb.com/search/title/?title_type=feature&year=2012-07-15,2012-07-21")
soup = BeautifulSoup(response.content,"html.parser")
```

```python
genres = soup.find_all("span",attrs={"class": "genre"})
runtime = soup.find_all("span",attrs={"class":"runtime"})
len(runtime),len(genres)
```

```python
genres[0].text.strip()
runtime[0].text
```

```python
film_types = []
runtime_genre=[]
for i in range(len(runtime)):
  runtime_genre.append((runtime[i].text.strip().split()[0],genres[i].text.strip().split(",")))
  film_types.extend(genres[i].text.strip().split(","))
```

```python
runtime_genre
```

```python
film_types
```

```python
for genre in set(film_types):
  print(genre,film_types.count(genre))
```

```python
from geopy.geocoders import Nominatim
```

```python
locator = Nominatim()
Srikakulam = locator.geocode("Srikakulam")
Ajmer = locator.geocode("Ajmer")
Wardha = locator.geocode("Wardha")
Alwar = locator.geocode("Alwar")
Sikar = locator.geocode("Sikar")

navi_mumbai = locator.geocode("Navi Mumbai")
kakinada = locator.geocode("Kakinada")
adoni = locator.geocode("Adoni")
jamnagar = locator.geocode("Jamnagar")
shivamogga = locator.geocode("Shivamogga")
sivasagar = locator.geocode("Sivasagar")
sagar = locator.geocode("Sagar")
# from geopy.distance import geodesic
# print(Their distance:,geodesic(kgp_lat_long,mad_lat_long).km)
```

```python
from geopy.distance import distance

print (f"Sikar - Navi Mumbai: {distance(Sikar[-1],navi_mumbai[-1]).km}")
print (f"Kakinada - Adoni: {distance(kakinada[-1],adoni[-1]).km}")
print (f"Jamnagar - Shivamogga: {distance(jamnagar[-1],shivamogga[-1]).km}")
print (f"Sivasagar - sagar: {distance(sivasagar[-1],sagar[-1]).km}")
```

```python
from geopy.distance import geodesic
print (f"Srikakulam - Alwar: {geodesic(Srikakulam.latitude,Alwar.latitude).km}")
print (f"Srikakulam - Ajmer: {geodesic(Srikakulam.latitude,Ajmer.latitude).km}")
print (f"Srikakulam - Wardha: {geodesic(Srikakulam.latitude,Wardha.latitude).km}")
print (f"Srikakulam - Sikar: {geodesic(Srikakulam.latitude,Sikar.latitude).km}")

print (f"Sikar - Navi Mumbai: {geodesic(Sikar.latitude,navi_mumbai.latitude).km}")
print (f"Kakinada - Adoni: {geodesic(kakinada.latitude,adoni.latitude).km}")
print (f"Jamnagar - Shivamogga: {geodesic(jamnagar.latitude,shivamogga.latitude).km}")
print (f"Sivasagar - sagar: {geodesic(sivasagar.latitude,sagar.latitude).km}")
```
